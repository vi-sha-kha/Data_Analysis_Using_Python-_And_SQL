```python
def income_data(inc):
    if inc == '<=50k':
        return 0
    elif inc == '>50k':
        return 1
    else:
        # Optionally handle unexpected values
        return None  # or raise an exception, depending on your needs
```

```python
data['encoded_income']=data['income'].apply(income_data)
```

```python
data.head(10)
```

```python
data.replace(to_replace=['<=50K', '>50K'], value=[0,1], inplace=True)
```
