```python
from io import IncrementalNewlineDecoder
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
##Display all the columns of the dataframe

pd.pandas.set_option("display.max_columns", None)
```

```python
dataset=pd.read_csv('train.csv')
print(dataset.shape)
```

```python
dataset.head()
```

```python
features_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>1]

for feature in features_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(), 4),  ' % missing values')
```

```python
for feature in features_with_na:
  data=dataset.copy()

  data[feature]=np.where(data[feature].isnull(),1,0)

  data.groupby(feature)['SalePrice'].median().plot.bar()
  plt.title(feature)
  plt.show()
```

```python
print("Id of Houses {}".format(len(dataset.Id)))
```

#Numerical variables

```python
#List of numerical variables
numerical_features=[feature for feature in dataset.columns if dataset[feature].dtypes!= 'O']

print('Number of numerical variables: ',len(numerical_features))

#visualize the numerical variables
dataset[numerical_features].head()
```

Temporal Variables(Eg: Datetime Variables)
From the dataset, we have 4 year variables.. We have to extract information from the datetime variables like no. of years or no. of days. Eg: Difference in years between the year the house was built and the year the house was sold. We will be performing the analysis in the Feature Engineering later on.

```python
#List of variables that contain year information
year_feature=[feature for feature in numerical_features if 'Year' in feature or 'Yr' in feature]

year_feature
```

```python
#Exploring the content of year varibles
for feature in year_feature:
  print(feature, dataset[feature].unique())
```

```python
#Analysing the temporal datetime variables
#To check whether there is a relation between the year the house is sold and Sale Price.

dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title("House Price vs Year Sold")
```

```python
# Difference between all the other year variable and the year the house was sold.

for feature in year_feature:
  if feature!='YrSold':
    data=dataset.copy()
    data[feature]=data['YrSold']-data[feature]

    plt.scatter(data[feature], data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.show()
```

```python
#Continuous variable and Discrete variables

discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature +['Id'] ]

print("Discrete Variables Count: {}".format(len(discrete_feature)))
```

```python
dataset[discrete_feature].head()
```

```python
#To find the relationship between discrete feature and Sale Price

for feature in discrete_feature:
  data=dataset.copy()
  data.groupby(feature)['SalePrice'].median().plot.bar()
  plt.xlabel(feature)
  plt.ylabel('Sale Price')
  plt.title(feature)
  plt.show()
```

```python
#Continuous Variable
continuous_feature=[feature for feature in numerical_features if feature not in discrete_feature + year_feature +['Id']]
print("Continuous feature Count: {}".format(len(continuous_feature)))
```

```python
for feature in continuous_feature:
  data=dataset.copy()
  data[feature].hist(bins=25)
  plt.xlabel(feature)
  plt.ylabel("Count")
  plt.title(feature)
  plt.show()
```

```python
#We will be using logarithmic transformation

for feature in continuous_feature:
  data=dataset.copy()
  if 0 in data[feature].unique():
    pass
  else:
    data[feature]=np.log(data[feature])
    data['SalePrice']=np.log(data['SalePrice'])
    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
## Outliers
for feature in continuous_feature:
  data=dataset.copy()
  if 0 in data[feature].unique():
    pass
  else:
    data[feature]=np.log(data[feature])
    data.boxplot(column=feature)
    plt.xlabel(feature)
    plt.ylabel(feature)
    plt.title(feature)
    plt.show()
```
